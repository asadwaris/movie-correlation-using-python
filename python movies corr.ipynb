```python
#scatterplot budget vs gross
plt.figure(figsize=(15,6))
sns_plot =sns.scatterplot(x='budget',y='gross',data=df)
fig = sns_plot.get_figure()
fig.savefig("budget vs gross.png")
```

```python
df.head()
```

```python
# regplot budget vs gross
plt.figure(figsize=(15,6))
sns_regplot =sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'purple'},line_kws={'color':'red'})
fig = sns_regplot.get_figure()
fig.savefig("reg plot budget vs gross.png")
```

```python
# visulize the correlation
corr=df.corr()
plt.figure(figsize=(15,6))
sns_heatmap =sns.heatmap(corr,annot=True)
fig = sns_heatmap.get_figure()
fig.savefig("correlation hear plot of movie data.png")
```
